````python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
````

````python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
````

````python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
````

````python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
````

## Defining the model

````python
class my_model():
    def __init__(self,X):
        self.weights = torch.rand(X.shape[1],1,dtype=torch.float64,requires_grad=True)
        self.bias = torch.zeros(1,dtype=torch.float64,requires_grad=True)
        
    def forward(self,X):
        z = torch.matmul(X,self.weights)+self.bias
        return torch.sigmoid(z)
    
    def loss_function(self,y_pred, y):
        # clamp predictions to avoid log(0)
        epsilon = 1e-7
        y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)
        loss =  -(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred)).mean()
        return loss
        
````

````python
learning_rate = 0.1
epochs = 25
````

## Training Pipeline 


### **1. Model Initialization**
```python
model = my_model(X_train_tensor)
```
- creating an instance of your model class, `my_model`. 
- Inside the `__init__` method, you initialize the model’s **parameters**:
  - `weights`: Random initial values with `requires_grad=True`.
  - `bias`: Zero-initialized with `requires_grad=True`.
- These parameters are trainable and will be updated during training.

---

### **2. Training Loop**
The loop iterates over a predefined number of `epochs`. Each epoch represents one full pass over the training data.

---

#### **Step 2.1: Forward Pass**
```python
y_pred = model.forward(X_train_tensor)
```
- The input data `X_train_tensor` is passed to the model’s `forward` method.
- Inside the forward method:
  - compute the linear combination: `z = X @ weights + bias` (matrix multiplication and bias addition).
  - Apply the **sigmoid activation** function: `torch.sigmoid(z)`, which gives predictions (`y_pred`) in the range `[0, 1]`.

---

#### **Step 2.2: Loss Calculation**
```python
loss = model.loss_function(y_pred, y_train_tensor)
```
- calculate the **binary cross-entropy loss**:
  - This measures the error between the predicted probabilities (`y_pred`) and the actual labels (`y_train_tensor`).
  - If the predictions are far from the true labels, the loss will be higher.
  - The loss is a scalar value that quantifies how well the model is performing.

---

#### **Step 2.3: Backward Pass**
```python
loss.backward()
```
- This is where **autograd** (PyTorch’s automatic differentiation engine) comes into play.
- The `loss.backward()` call:
  - Computes the **gradients of the loss** with respect to the model’s parameters (`weights` and `bias`).
  - These gradients are stored in `model.weights.grad` and `model.bias.grad`.
  - The gradients represent how much each parameter contributes to the loss, and they are used to update the parameters in the next step.

---

#### **Step 2.4: Parameter Updates**
```python
with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad
```
- The parameters (`weights` and `bias`) are updated using **gradient descent**:
  - Subtract the product of the gradient (`.grad`) and the learning rate (`learning_rate`).
- **Why `torch.no_grad()`?**
  - This ensures that the update operation (the subtraction) is not tracked in the computation graph, as these updates are not part of the backpropagation process. 
  - It saves memory and computation time.

---

#### **Step 2.5: Clear Gradients**
```python
model.weights.grad.zero_()
model.bias.grad.zero_()
```
- After updating the parameters, the gradients (`.grad`) are **cleared**.
- **Why?**
  - Gradients accumulate in PyTorch by default. If you don’t clear them, gradients from the previous epoch will be added to the new gradients in the next epoch, which will result in incorrect updates.
  - Clearing gradients ensures that only the gradients for the current epoch are used.

---

### **3. Output**
```python
print(f'Epoch {epoch+1}, Loss: {loss.item()}')
```
- At the end of each epoch, print the loss value.
- This allows you to monitor how the model is improving (or not) over time.

---

### **Summary of the Flow**
1. **Forward Pass**:
   - Use input data to calculate predictions (`y_pred`).
   - Compute the loss based on how far predictions are from true labels.

2. **Backward Pass**:
   - Calculate the gradients of the loss with respect to the model’s parameters.

3. **Parameter Update**:
   - Adjust the parameters (`weights` and `bias`) using gradient descent to minimize the loss.

4. **Clear Gradients**:
   - Reset gradients for the next epoch to avoid accumulation.

This loop continues for the specified number of epochs, gradually improving the model’s predictions by iteratively minimizing the loss.

````python
# create model
model = my_model(X_train_tensor)

# define loop
for epoch in range(epochs):
    # forward pass
    y_pred = model.forward(X_train_tensor)
    
    # loss calculate 
    loss = model.loss_function(y_pred, y_train_tensor)
    print(f'Epoch {epoch+1}, Loss: {loss.item()}')
    
    # backward pass
    loss.backward()
    
    # parameters update 
    with torch.no_grad():
        model.weights -= learning_rate*model.weights.grad
        model.bias -= learning_rate*model.bias.grad
        
    # clear the gradients 
    model.weights.grad.zero_()
    model.bias.grad.zero_()
    
    # print loss in each epoch
    print(f'Epoch {epoch+1}, Loss: {loss.item()}')
````

````python
with torch.no_grad():
    y_pred = model.forward(X_test_tensor)
    y_pred = (y_pred > 0.9).float()
    accuracy = (y_pred == y_test_tensor).float().mean()
    print(f'Accuracy: {accuracy}')
    
````
